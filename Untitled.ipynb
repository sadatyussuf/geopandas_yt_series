```python
raw_data = pd.read_csv('traffic_data.csv')
data = raw_data[raw_data['Segment'] != 'Full Route']
full_route_data = raw_data[raw_data['Segment'] == 'Full Route']
```

```python
sns.lineplot(data=full_route_data, x='Date', y='Duration', hue='Route_Direction')
plt.show()
```

```python
# Convert duration to seconds for easier calculation
def convert_duration_to_seconds(duration):
    min_sec = duration.split(':')
    return int(min_sec[0]) * 60 + float(min_sec[1])

raw_data['Duration_Seconds'] = raw_data['Duration'].apply(convert_duration_to_seconds)

# Plot average duration by mode of transport
sns.barplot(x='Mode_of_Transport', y='Duration_Seconds', data=raw_data, estimator=lambda x: sum(x)/len(x))
plt.title('Average Duration by Mode of Transport')
plt.ylabel('Average Duration (Seconds)')
plt.show()
```

```python
# Convert Date to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Plot travel time trends across dates for in-run and outrun
sns.lineplot(x='Date', y='Duration_Seconds', hue='Route_Direction', data=data)
plt.title('Travel Time Trends Across Dates')
plt.ylabel('Duration (Seconds)')
plt.show()
```

```python
# Boxplot to show the distribution of durations for in-run and outrun
sns.boxplot(x='Route_Direction', y='Duration_Seconds', data=raw_data)
plt.title('In-Run vs. Outrun Duration Comparison')
plt.ylabel('Duration (Seconds)')
plt.show()
```

```python
# Plot segment durations (Shangrila to Airport 2nd and Airport 2nd to Shangrila)
sns.barplot(x='Segment', y='Duration_Seconds', hue='Route_Direction', data=raw_data)
plt.title('Segment Duration Variability')
plt.ylabel('Duration (Seconds)')
plt.show()
```
